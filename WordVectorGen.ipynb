```python
import pandas as pd
```

```python
df = pd.read_csv('tweet_clean.csv')
```

```python
df.head()
```

```python
clean = [x.split() for x in df.cleaned.astype('str')]
```

```python
clean
```

```python
from gensim.models.fasttext import FastText
model = FastText(window=4, min_count=1) 
model.build_vocab(sentences=clean)
model.train(sentences=clean, total_examples=len(clean), epochs=150)
```
